```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("customer_support_tickets.csv")
```

```python
df.info()
```

<h3>Exploratory Data Analysis</h3>

```python
frequent_issues = df['Ticket Subject'].value_counts()

plt.figure(figsize=(8,6))
frequent_issues.plot(kind='bar')
plt.title('frequent issues')
```

```python
ticket_status = df['Ticket Status'].value_counts()

plt.figure(figsize=(8,6))
ticket_status.plot(kind='pie')
plt.title('frequent issues')
```

```python
plt.figure(figsize=(10, 6))
product_purchased_distribution = df['Product Purchased'].value_counts().head(15)
product_purchased_distribution.plot(kind='barh')
plt.title('Top 10 Products Purchased')
plt.xlabel('Count')
```

```python
ticket_types = df['Ticket Type'].value_counts()
plt.figure(figsize=(8,6))
ticket_types.plot(kind='pie',autopct='%1.1f%%')
plt.title("Ticket Types Distribution")
```

<h3>average rating by user based on ticket type</h3>

```python
import matplotlib.pyplot as plt
import seaborn as sns


df_filtered = df.dropna(subset=["Customer Satisfaction Rating"])


rating_by_ticket_type = df_filtered.groupby("Ticket Type")["Customer Satisfaction Rating"].mean().sort_values()

plt.figure(figsize=(10, 5))
sns.barplot(x=rating_by_ticket_type.index, y=rating_by_ticket_type.values, palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Ticket Type")
plt.ylabel("Average Customer Satisfaction Rating")
plt.title("Customer Satisfaction Rating by Ticket Type")
plt.show()


rating_by_ticket_type
```

<h3>average rating by user based on gender</h3>

```python
df_filtered = df.dropna(subset=["Customer Satisfaction Rating"])


rating_by_ticket_gender = df_filtered.groupby("Customer Gender")["Customer Satisfaction Rating"].mean().sort_values()

plt.figure(figsize=(10, 5))
sns.barplot(x=rating_by_ticket_gender.index, y=rating_by_ticket_gender.values, palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Ticket Type")
plt.ylabel("Average Customer Satisfaction Rating")
plt.title("Customer Satisfaction Rating by Ticket Channel")
plt.show()


rating_by_ticket_gender
```

```python
# Define age bins and labels
bins = [18, 25, 35, 45, 55, 65, 75]  # Age groups
labels = ["18-25", "26-35", "36-45", "46-55", "56-65", "66-75"]

# Apply binning
df_filtered["Age Group"] = pd.cut(df_filtered["Customer Age"], bins=bins, labels=labels, right=False)

# Compute average satisfaction rating per age group
rating_by_age_group = df_filtered.groupby("Age Group")["Customer Satisfaction Rating"].mean()

# Plot the results
plt.figure(figsize=(10, 5))
sns.barplot(x=rating_by_age_group.index, y=rating_by_age_group.values, palette="viridis")
plt.xlabel("Age Group")
plt.ylabel("Average Customer Satisfaction Rating")
plt.title("Customer Satisfaction Rating by Age Group")
plt.show()

# Display computed values
rating_by_age_group
```
